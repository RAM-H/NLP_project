```python
import requests
from bs4 import BeautifulSoup as soup 
```

The Plant List was a working list of all known plant species that aimed to be comprehensive for species of Vascular plant (flowering plants, conifers, ferns and their allies) and of Bryophytes (mosses and liverworts). we did use this website to scrap all the species names .

```python
html =requests.get("http://www.theplantlist.org/1.1/browse/-/-/")
html.status_code
```

```python
bsobj = soup(html.content,'lxml')
```

```python
genus=bsobj.find('ul',{'id':'nametree'})
```

```python
genus.find_all('a')
```

```python
genus=genus.find_all('a')
```

From the HTML code of a page we can extract all the links of the species ,and store them to be used in further search

```python
links=[]
for g in genus:
    if g.find_all('i',class_="Accepted genus"):
       links.append("http://www.theplantlist.org"+g['href'])
```

The next step is get the species names from all the pages in the plant list , the name of the specie is composed of two parts ,the genus and the species

```python
scrapped_species=[]
for link in links:
    html =requests.get(link)
    bsobj = soup(html.content,'lxml')
    try:
        species=bsobj.find('tbody')
        species=species.find_all('td',class_='name Accepted')
    except Exception as e:
        print(link)
        continue
    for s in species:
        scrapped_species.append(s.find('i',class_='genus').text+' '+s.find('i',class_='species').text)
```

The Requested URLs above were not found on the server ,and we bypassed them by a 'try except' 

```python
len(scrapped_species)
```

```python
with open('../genereted data/species-THEPLANTLIST.txt', 'w') as f:
    for es in scrapped_species:
        f.write(es)
        f.write('\n')
```
